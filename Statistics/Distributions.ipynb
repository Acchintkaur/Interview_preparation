Importing Libraries

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm
```

Mean

```python
def find_mean(df):
    return sum(df)/len(df)
```

Variance

```python
def find_var(df):
    mean_cal = find_mean(df)
    
    diff_sq = []
    for i in df:
        diff_sq.append((i-mean_cal)**2)
    return find_mean(diff_sq)
```

Standard Deviation

```python
def find_std(df):
    var = find_var(df)
    
    return var**0.5
```

Normal distribution

```python
def normal_dist(df):
    pdf = (np.pi*find_std(df))*np.exp(-0.5*((df - find_mean(df))/find_std(df))**2)
    return pdf

# plotting the graph
def plotting_normal(df,pdf):
    plt.plot(df,pdf,color = 'red')
    plt.xlabel('Data points')
    plt.ylabel('Probability Density')
```

Driver Code

```python
df = np.linspace(1,50,200)
pdf = normal_dist(df)
# print(pdf)
print("Mean : ",find_mean(df))
print("Variance : ", find_var(df))
print("standard deviation : ", find_std(df))
plotting_normal(df,pdf)
```

Without code

```python
pdf_withoutc = norm.pdf(df,loc =find_mean(df),scale =find_std(df))
plotting_normal(df,pdf_withoutc)
```
